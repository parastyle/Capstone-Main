```python
import time
import random

subscription_key = '##########################'
assert subscription_key
search_url = "https://api.cognitive.microsoft.com/bing/v7.0/news/search"
search_term = "ford employees january"
x_search = '199.99.99.99'
import requests


```

```python


    
headers = {"Ocp-Apim-Subscription-Key" : subscription_key}
params  = {"q": search_term, "textDecorations": True, "textFormat": "raw",'count':100,'offset':7800,'sortBy':'relevance'}

response = requests.get(search_url, headers=headers, params=params)
response.raise_for_status()
search_results = response.json()
descriptions = [article["description"] for article in search_results["value"]]
```

```python

term_append = ['january', 'february', 'march', 'april', 'may', 'june', 'july', 'august', 'september', 'october', 'november' , 'december',
               'motors', 'automotive', 'cars' , 'recall', 'production', 'technology', 'issue' , 'publicity' , 'price' , 
               'competition' , 'horsepower', 'domestic', 'international', 'concept', 'hybrid', 'global', 'employees' , 
               ' shareholders', 'stock', 'auto,','j.d', 'manufacturing','horsepower'] 
print(len(term_append))
```

```python
import pandas as pd
loopCount = 0



#mdf = pd.read_csv(search_term.replace(' ','_')+'.csv').drop('Unnamed: 0',axis=1)
smdf = pd.DataFrame()

search_term = 'Mercedes'
for term in term_append:
    search_term_appended = search_term+' ' + term
    print(search_term_appended)
    mdf = pd.DataFrame()
    offsetCount = 0
    for counter in range(10):
        loopCount += 1
        if counter % 2 == 0:
            time.sleep(random.randint(1,2))
            print('sleeping at offsetCount = ' + str(offsetCount) + ' and request '+str(counter))
            mdf.to_csv(search_term_appended.replace(' ','_')+'.csv')
        try:
            params  = {"q": search_term_appended,'X-Search-ClientIP': x_search, "textDecorations": True, "textFormat": "HTML",'count':100,'offset':offsetCount,'sortBy':'Date'}

            response = requests.get(search_url, headers=headers, params=params)
            response.raise_for_status()
            search_results = response.json()
            desc,head,url,provider,date = [],[],[],[],[]
            for x in search_results['value']:
                desc.append(x['description'])
                head.append(x['name'])
                date.append(x['datePublished'])
                url.append(x['url'])
                provider.append(x['provider'][0]['name'])
            obs = pd.DataFrame({'description':desc,'headline':head,'url':url,'provider':provider,'date':date})
            mdf = mdf.append(obs)
            smdf = smdf.append(obs)
        except:
            print('failure at '+str(counter) + 'offset = ' + str(offsetCount))
            break
        offsetCount += 100
    

smdf.to_csv(search_term_appended.replace(' ','_')+'.csv',index=False)
```
